```python

# -*- coding: utf-8 -*-
"""
Created on Fri Apr  1 11:15:39 2022

@author: Jishnu N
"""

# IMPORTING REQUIRED LIBRARIES

import pandas as pd
import numpy as np
from tqdm import tqdm
from time import sleep

# READING INPUT FILE

data = pd.read_csv(r'den_caca500.out')
df = pd.DataFrame(data)
data.to_csv('blah.csv')  # converting to csv file

# changing the FORMATTING of input file into new file (no change in data)

raw = []
with open('blah.csv', 'r') as f:
    for line in f:
        raw.append(line.split())
data1 = pd.DataFrame(raw)
data_ind = data1.set_index
data1.columns = ['S_no', 'x', 'y', 'z', 'Px', 'Py', 'Pz', 'mass', 'particle_index', 'collision_no']
data1_cols = data1[['x', 'y', 'z', 'Px', 'Py', 'Pz', 'particle_index']]

data1_cols = data1_cols.astype(float)
no = int(input("Enter atomic number :"))
t = int(input("Enter No of Timestamps to display :"))  # No. of timeintervals to display
z = 4

# selecting desired data from whole data and storing it in 'array'
u = []

for index in tqdm(range(t)):

    for row_series in data1_cols.iterrows():
        x2 = z
        y2 = x2 + no
        data1_main = data1_cols.iloc[x2:y2]
        data1_main = data1_main.astype(float)

    u.append(data1_main)
    z = z + (no + 2)
    sleep(0.8)

array = np.asarray(u, dtype=float, order=None)
# print(array)


dff = pd.DataFrame(array.reshape(-1, 7))  # changing array into pandas dataframe for ease of working
dff.columns = ['x2', 'y2', 'z2', 'Px2', 'Py2', 'Pz2', 'particle_index1']

# x1 = dff['x2']
# y1 = dff['y2']
# z1 = dff['z2']
# px = dff['Px2']
# py = dff['Py2']
# pz = dff['Pz2']
# ind = dff['particle_index1']
#
# array1 = np.column_stack((x1, y1, z1, px, py, pz, ind))
# print(array1)
# 'array1' contains whole set of position cordinates of nucleons



# WRITING ELECTRIC FIELD EQUATION

# print(array1)
# P = np.sqrt((px ** 2) + (py ** 2) + (pz ** 2))



onp = dff.loc[dff['particle_index1'] == 2]
onp.columns = ['x3', 'y3', 'z3', 'Px3', 'Py3', 'Pz3', 'index1']
onp= onp.reset_index(drop = True)
onpar= onp.to_numpy(dtype=float)
# print(onpar)
# print(onp)


#
# #print(eqn)
m=0
p=[]
for index in tqdm(range(t)):
    row1 = onpar[m:(m + 40), :]
    #print(row1)
    x1 = row1[:, 0]
    # print(x1)
    y1 = row1[:, 1]
    z1 = row1[:, 2]
    px = row1[:, 3]
    py = row1[:, 4]
    pz = row1[:, 5]
    c = 2.998 * (10 ** 8)
    vx = (px / 0.938)*c
    vy = (py / 0.938)*c
    vz = (pz / 0.938)*c
    vel = np.sqrt((vx ** 2) + (vy ** 2) + (vz ** 2))
    rnvn = (x1*(10**-15) * vx) + (y1*(10**-15) * vy) + (z1*(10**-15) * vz)
    # print(vel)
    pos = (np.sqrt((x1 ** 2) + (y1 ** 2) + (z1 ** 2)))*(10**-15)
    denom = (c * pos - rnvn)
    xyz=np.ndarray(shape=(0))
    # print(denom)

    for element in denom:
        if 0.0 < element<0.00001 :
            xyz=np.append(xyz, [element])
        else:
            xyz=np.append(xyz, [1*(10**-5)])

    # arxyz = np.asarray(xyz, dtype=float, order=None)
     # rhotrim = np.asarray(f, dtype = float, order = None)
    eqx = (c * x1 - (pos * vx)) / (xyz ** 3)
    eqy = (c*y1 - (pos * vy)) / ((c*pos - rnvn) ** 3)
    eqz = (c*z1 - (pos * vz)) / ((c*pos - rnvn ) ** 3)
    #eq1 = np.sqrt((eqx ** 2) + (eqy ** 2) + (eqz ** 2))
    eq2 = c**2 - (vel ** 2)
    const = (1.6022*(10**-19)) / (4 * np.pi * (8.854*(10**-12)))
    eqn = eqx * eq2

    totalsm = sum(eqn)
    total=(totalsm*const)
    # print(total)
    p.append(total)
    m = m + 40
    sleep(0.8)
arr = np.asarray(p, dtype=float, order=None)
# print(arr)
ni100 = pd.DataFrame(arr.reshape(-1, 10))
ni100.columns = ['1fm/c', '5fm/c', '10fm/c', '15fm/c', '20fm/c', '30fm/c', '50fm/c', '100fm/c', '150fm/c', '200fm/c']
cols100 = ['1fm/c', '5fm/c', '10fm/c', '15fm/c', '20fm/c', '30fm/c', '50fm/c', '100fm/c', '150fm/c', '200fm/c']
meen = ni100[cols100].mean()
print(meen)

# with open('out_ni_may9_eMag.txt', 'w') as k:
#     print(meen, file=k)
```
